```python
def generate_random_stuff(seed = None):
    random.seed(seed)
    results = []
    for _ in range(5):
        results.append(random.randint(0, 5))
    characters = list('abc')
    random.shuffle(characters)
    results.append(characters)
    for _ in range(5):
        results.append(random.gauss(0, 1))
    return results
```

```python
from collections import namedtuple
Freq = namedtuple('Freq', 'count freq')
```

```python
def freq_counts(lst):
    total = len(lst)
    return {k: Freq(lst.count(k), 100 * lst.count(k) / total) for k in set(lst)}
```

```python
freq_counts(random.choices(l, k=100_000))
```

```python
freq_counts(random.choices(l, k=100_000, cum_weights = cum_weights))
```

```python
import random
random.seed(0)
demos = random.choices([0,1], k=10_000_000, weights = [1,9])
```

```python
start = time.perf_counter()
for d in demos:
    if d == 0:
        continue
    else:
        10/ d
end = time.perf_counter()
print(f'Avg Elapsed Time: {(end - start)/len(demos)}')
```

```python
demos = random.choices([0,1], k=10_000_000, weights = [9, 1])
start = time.perf_counter()
for d in demos:
    try:
        10/d
    except ZeroDivisionError:
        pass
end = time.perf_counter()
print(f'Avg Elapsed Time: {(end - start)/len(demos)}') # code is slower when the exceptions are much more
```

don't use * args and **kwargs blindly : instead give meaningful names

```python
def audit(func):
    def inner(*args, **kwargs):
        print(f'called {func.__name__}')
        return func(*args, **kwargs)
    return inner
```

```python
@audit
def say_hello(name):
    return f'Hello {name}'
from operator import mul
from functools import reduce
@audit
def product(*values):
    return reduce(mul, values)  
```

```python
say_hello('Polly')
```

```python
def audit(func):
    def inner(*args, **kwargs):
        print(f'called {func.__name__}')
        return func(*args, **kwargs)
    return inner
```

```python
audit(say_hello)('Polly')
```

```python
class Person:
    def __init__(self, name, age, **custom_attributes: 'this is a meaningful name'):
        self.name = name
        self.age = age
        for attr_name, attr_value in custom_attributes.items():
            setattr(self, attr_name, attr_value)
```

```python
parrot = Person('Polly', 101, status = 'stiff', vooms = False)
print(vars(parrot))
```

Sentinel values for parameter defaults: sentinel values are used to distinguish between none and no value provided

```python
def validate(a = None):
    if a is not None:
        print('argument was provided')
    else:
        print('argument was not provided')
```

```python
validate(a = None)
```

```python
validate( None)
```

```python
_sentinel = object()
```

```python
def validate(a = _sentinel):
    if a is not _sentinel:
        print('argument was provided')
    else:
        print('argument was not provided')
```

```python
validate(None)
```

```python
def validate(a = object()):
    default_a = validate.__defaults__[0]
    if a is not default_a:
        print('argument was provided')
    else:
        print('argument was not provided')
```

```python
validate(10)
```

```python
validate()
```

```python
validate(None)
```

```python
def validate(a = object(), b=object(),*, kw = object()):
    default_a = validate.__defaults__[0]
    default_b = validate.__defaults__[1]
    default_kw = validate.__kwdefaults__['kw']
    print(default_a is default_b)
    print(default_b is default_kw)
    if a is not default_a:
        print('argument a was provided')
    else:
        print('argument a was NOT provided')
    if b is not default_b:
        print('argument b was provided')
    else:
        print('argument b was NOT provided')
    if kw is not default_kw:
        print('argument kw was provided')
    else:
        print('argument kw was NOT provided')
```

```python
validate(100, 200, kw = None)
```

```python
validate(200)
```
