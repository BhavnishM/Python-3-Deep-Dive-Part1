Reducing Functions

```python
l = [5, 8, 6, 10, 9]
```

```python
_max = lambda x, y: x if x>y else y
```

```python
def max_sequence(sequence):
    result = sequence[0]
    for x in sequence[1:]:
        result = _max(result, x)
    return result
```

```python
max_sequence(l)
```

```python
_min = lambda x, y: x if x<y else y
```

```python
def min_sequence(sequence):
    result = sequence[0]
    for x in sequence[1:]:
        result = _min(result, x)
    return result
```

```python
min_sequence(l)
```

```python
_add = lambda x, y: x + y
```

```python
def add_sequence(sequence):
    result = sequence[0]
    for x in sequence[1:]:
        result = _add(result, x)
    return result
```

```python
add_sequence(l)
```

```python
def reduce(fn, sequence):
    result = sequence[0]
    for x in sequence[1:]:
        result = fn(result, x)
    return result
```

```python
reduce(_add, l)
```

```python
reduce(_max, l)
```

```python
reduce(_min, l)
```
