```python
def from_base10(n, b):
    '''
    if b < 2:
        raise ValueError('Base b must be >= 2')
    if n < 0:
        raise ValueError('Number n must be >= 0')'''
    if n == 0:
        return[0]
    digits = []
    while n>0:
        n, m = divmod(n, b)
        digits.insert(0,m)
    return digits
```

```python
from_base10(10,2)
```

```python
from_base10(255,16)
```

```python
def encode(digits, digits_map):
    if max(digits) >= len(digits_map):
        raise ValueError('digit_map is not long enough to encode the digits')
    return ''.join([digits_map[d] for d in digits])
```

```python
encode([15, 15], '0123456789ABCDEF')
```

```python
def rebase_from10(n, base):
    if base <2 or base > 36:
        raise ValueError('Base b must be between 2 and 36')
    sign = -1 if n < 0 else 1
    n *= sign
    digits_map = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    digits = from_base10(n, base)
    encoding = encode(digits, digits_map)
    if sign == -1:
        encoding = '-' + encoding
    return encoding
```

```python
rebase_from10(123456789,35)
```

```python
rebase_from10(-123456789,35)
```

```python
from numba import njit 
@njit
def njit_manual(a):
    real = a.real
    imag = a.imag
    #r = real**2 + a.imag**2
    r = math.sqrt(math.hypot(real, imag))
    return complex(math.sqrt((r+real)/2) , math.copysign(math.sqrt((r-real)/2), imag))
```

```python
import time
n = 100000000
```

```python
n = 100000000
def time_it(a, n, f):
    for i in range(n):
        f(a)
start = time.perf_counter()
time_it(a, n, cmath.sqrt)
end = time.perf_counter()
start - end
```

```python
start = time.perf_counter()
time_it(a, n, njit_manual)
end = time.perf_counter()
start - end
```

```python
%%cython
import math
cpdef complex cython_manual(complex a):
    real = a.real
    imag = a.imag
    #r = real**2 + a.imag**2
    cdef double r = (real**2 + imag** 2)**0.5/2
    cdef double im = real/2
    cdef double real_part = (r + im)**0.5
    cdef double imag_part = math.copysign((r-im)**0.5, imag)
    return real_part + imag_part*1j
```

```python
import time
start = time.perf_counter()
time_it(a, n, cython_manual)
end = time.perf_counter()
start - end
```
